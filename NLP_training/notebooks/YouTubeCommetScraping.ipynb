```python
import googleapiclient.discovery
import pandas as pd

api_service_name = "youtube"
api_version = "v3"
DEVELOPER_KEY = "AIzaSyAmBkakp9o3OuZgpjNzSrRIPR3iMF5y7nQ"
```

```python
comments = []
```

```python
youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey = DEVELOPER_KEY)
```

```python
def fill_comment_by_YT(videoIdINPUT):
  request = youtube.commentThreads().list(
      part="snippet",
      videoId= videoIdINPUT,
      maxResults=100
  )


  response = request.execute()

  for item in response['items']:
    comment = item['snippet']['topLevelComment']['snippet']
    public = item['snippet']['isPublic']
    comments.append([
        comment['textOriginal'],
        public
    ])

  while (1):
    try:
      nextPageToken = response['nextPageToken']
    except KeyError:
      break
    nextPageToken = response['nextPageToken']

    nextRequest = youtube.commentThreads().list(
      part="snippet",
      videoId="ft9MTPDwrHs",
      maxResults=100,
      pageToken=nextPageToken
    )

    response = nextRequest.execute()

    for item in response['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      public = item['snippet']['isPublic']
      comments.append([
        comment['textOriginal'],
        public
      ])
```

```python
video_ids = [
    "6Iz2xLNSxwE",
    "v3G2kPhq77c",
    "Iaq0feB8pqI",
    "7VbzL_ONE0k",
    "kMkYRkEAsaY",
    "Gwz8ug8ZvgU",
    "OpEceC7bb9o"
]
for id in video_ids:
  fill_comment_by_YT(id)
```

```python
df = pd.DataFrame(comments, columns=['comment' , 'public'])
df.info()
```

```python
df = df[df['comment'].str.len() >= 10]
```

```python
import emoji

def remove_emoji(text):
  return emoji.replace_emoji(text, replace='')
```

```python
df['comment'] = df['comment'].apply(remove_emoji)
df.head()
```

```python
df = df[df['comment'].str.len() > 10]
df.shape
```

```python
def remove_p (text):
  text = text.replace('.', '')
  text = text.replace(',', '')
  text = text.replace('!', '')
  text = text.replace('?', '')
  text = text.replace(':', '')
  text = text.replace(';', '')
  return text

df['comment'] = df['comment'].apply(remove_p)
```
